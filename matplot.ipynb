```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2, 5, 6]  
y = [4, 10, 4]  
plt.plot(x, y) # default line 
plt.title('Line graph')  
plt.ylabel('Y axis')  
plt.xlabel('X axis')
plt.xticks(list(range(0,9,2)))
plt.show()  # to remove default information
```

```python
# plot :from array
import matplotlib.pyplot as plt 
import numpy as np
  
x = np.arange(1,11)
y = np.arange(12,22)  
plt.plot(x, y,color="#FFD700",linewidth=5,linestyle="-.") 
plt.title('Line graph')  
plt.ylabel('Y axis')  
plt.xlabel('X axis')  
plt.show()
```

```python
# plot : scatter
import matplotlib.pyplot as plt 
import numpy as np
  
x = np.arange(1,11)
y = np.arange(12,22)  
plt.scatter(x, y) # default line 
plt.title('scatter Plot')  
plt.ylabel('Y axis')  
plt.xlabel('X axis')  

plt.show()  
```

```python
#plot : pass the arbitrary number of arguments to the plot(). 
from matplotlib import pyplot as plt  
plt.figure(figsize=(4,3.5))
plt.plot([1,2,3,4,5],[1,4,9,16,25],marker='*')  
plt.ylabel("y axis")  
plt.xlabel('x axis')  
plt.show()  
```

```python
# plot a bar graph by using year as per online_sale
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.bar(year,online_sale,color="black",)
plt.xticks(rotation=45)
plt.title("year wise sale comparisions")
plt.xlabel("years")
plt.ylabel("sales in crores")
plt.show()

```

```python
# customizing color and width of bar  and adding title,lables in above eg:
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,3.3))
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.bar(langs,students,color ='green',width = 0.5)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```
